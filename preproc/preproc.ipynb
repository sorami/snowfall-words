# アンケートデータ


```python
import pandas as pd
```

## Load


### Data （事前収集）


```python
df = pd.read_csv("data.csv")
df.head()
```

```python
sentences = []

for index, row in df.iterrows():
    sent = row[1:].tolist()
    sentences.append(sent)

sentences
```

```python
len(sentences)
```

### Data（会期中収集）


```python
df2 = pd.read_csv("./data2.csv", header=None)
df2
```

```python
sentences2 = []
for index, row in df2.iterrows():
    sent = [d for d in row.tolist() if d == d]  # remove nan
    sentences2.append(sent)

sentences2
```

```python
len(sentences2)
```

```python
sentences_all = sentences + sentences2
len(sentences_all)
```

## Normalize


### 半角 → 全角


```python
def detect_ascii_chars(sentences):
    ascii_chars = []
    for sent in sentences:
        for clause in sent:
            ascii_chars.extend([c for c in clause if c.isascii()])
    return ascii_chars
```

```python
set(detect_ascii_chars(sentences_all))
```

```python
def half_to_full_width_char(s):
    trans = str.maketrans("0123456789()GLA&", "０１２３４５６７８９（）ＧＬＡ＆")
    return s.translate(trans)


for i, sent in enumerate(sentences_all):
    for j, clause_before in enumerate(sent):
        clause_after = "".join([half_to_full_width_char(c) for c in clause_before])
        if clause_before != clause_after:
            print(clause_before, "=>", clause_after)
            sentences_all[i][j] = clause_after
```

```python
set(detect_ascii_chars(sentences_all))
```

### 句読点


```python
for i, sent in enumerate(sentences_all):
    for j, clause in enumerate(sent):
        puncts = [c for c in clause if c in ".,!?。、！？"]
        if puncts:
            print(puncts, clause)
```

```python
# 末尾の句読点を削除
for i, sent in enumerate(sentences):
    for j, clause in enumerate(sent):
        new_clause = clause.rstrip(".,?。、？")
        if new_clause != clause:
            print(clause, "=>", new_clause)
            sentences[i][j] = new_clause
```

## 出力


```python
with open("sentences.txt", "w") as f:
    for sent in sentences_all:
        print(sent, end=",\n", file=f)
```
